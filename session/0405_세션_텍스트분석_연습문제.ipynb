```python
import nltk
nltk.download("book", quiet=True)
from nltk.book import *
```

```python
# 파일 목록 확인
nltk.corpus.gutenberg.fileids()
```

```python
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(emma[:5000])
```

### (1) 문제 1 : 단어 토큰화를 진행하고 토큰의 개수와 중복되지 않는 단어의 수를 확인해보세요.

```python
from nltk import word_tokenize
words = word_tokenize(emma)
print(words)
```

```python
len(words)
# 191785 개의 토큰이 있다.
```

```python
len(list(set(words)))
# 8406 개의 단어가 있다.
```

### (2) 문제 2 : stop words를 제거하고 남아있는 토큰의 개수와 중복되지 않는 단어는 몇개 남았는지 확인해보세요.

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')

filtered_words=[]
stop_words = []

for word in words:
  word = word.lower()
  if word not in stopwords:
    filtered_words.append(word)
  else:
    stop_words.append(word)

print(filtered_words)
print(stop_words)
```

```python
len(filtered_words) # 남아있는 토큰 수
```

```python
len(stop_words) # 제거된 토큰 수
```

```python
len(list(set(filtered_words))) # 남아있는 단어 수(중복 x)
```

```python
# 추가
len(list(set(stop_words))) # 제거된 stop words 수(중복x)
```
