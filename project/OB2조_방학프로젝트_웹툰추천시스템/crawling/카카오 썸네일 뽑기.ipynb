```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from PIL import Image
import time


options = Options()
user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36"
options.add_argument('user-agent=' + user_agent)
options.add_argument("lang=ko_KR")
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
options.add_argument("--no-sandbox")
driver = webdriver.Chrome(options=options)


for index, url in enumerate(kakao['Link'][1357:]):
        
    driver.get(url)
    time.sleep(3) 

    # 로그인 페이지 확인 및 로그인 처리
    if "로그인이 필요합니다" in driver.page_source:
        kakao_id = KAKAO_ID # 환경 변수에서 카카오 ID 가져오기
        kakao_pw = KAKAO_PW # 환경 변수에서 카카오 PW 가져오기
        login_button = driver.find_element(By.CSS_SELECTOR, 'button.relative.px-10.py-0.btn-transparent-grey-06')
        login_button.click()  # '카카오계정 직접 입력' 버튼 클릭
        time.sleep(2)
        driver.find_element(By.ID, 'loginId--1').send_keys(kakao_id)
        driver.find_element(By.ID, 'password--2').send_keys(kakao_pw)
        driver.find_element(By.CSS_SELECTOR, 'button.btn_g.highlight.submit').click()

        time.sleep(5)  # 로그인 후 대기
        driver.implicitly_wait(10)


    # 요소 찾기 (해당 XPath로 변경)
    element = driver.find_element(By.XPATH, '//*[@id="character-container"]/div')

    # 요소의 위치와 크기 얻기
    location = element.location
    size = element.size

    # 전체 페이지 스크린샷 찍기
    driver.save_screenshot(f'img_kakao/full_{index+1357}.png')

    # 전체 스크린샷 열기
    image = Image.open(f'img_kakao/full_{index+1357}.png')

    # 요소의 위치와 크기를 기반으로 이미지 크롭
    left = location['x']
    top = location['y']
    right = location['x'] + size['width']
    bottom = location['y'] + size['height']

    cropped_image = image.crop((left, top, right, bottom))

    # 크롭된 이미지 저장
    cropped_image.save(f'img_kakao/{index+1357}.png')

driver.quit()
```
