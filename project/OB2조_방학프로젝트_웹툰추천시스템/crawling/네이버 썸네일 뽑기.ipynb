```python
driver = webdriver.Chrome()

driver.get('http://nid.naver.com/nidlogin.login')
time.sleep(20)

naver_img_url = []

for url in naver['Link']:
    driver.get(url)
    driver.implicitly_wait(10)
    image_elements = driver.find_elements(By.XPATH, '//*[@id="content"]/div[1]/button/div/img')
    if image_elements:
        image_url = image_elements[0].get_attribute('src')
        
    naver_img_url.append(image_url)
    time.sleep(1)
    
driver.quit()
```
