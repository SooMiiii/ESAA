```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def get_webtoon_links():
    url = 'https://webtoon.kakao.com/'
    
    link_list = []
    day_list = []
    rank_list = []
    
    driver = webdriver.Chrome()
    

    ## (1) 요일별로 링크, 순위 가져오기
    for day in ['mon','tue','wed','thu','fri','sat','sun','complete']:
        day_url = f'{url}?tab={day}'
        driver.get(day_url)
        driver.implicitly_wait(10) 
        time.sleep(1)

        # 버튼 클릭하기 전에 정확한 요소 확인
        wait = WebDriverWait(driver, 20)  # 대기 시간을 20초로 증가
        button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#root > main > div > div.px-11.mx-auto.my-0.w-full.lg\\:w-default-max-width.md\\:w-\\[490px\\] > div.sticky.w-full.z-navigationBarPlus1.mt-6.transition-all.duration-\\[250ms\\].bg-background > div.w-full.h-40.sticky.flex.px-11.border-t-1.pb-10.justify-center.overflow-hidden > div > button:nth-child(1)')))

        # 버튼 클릭
        button.click()
        print("버튼이 클릭되었습니다.")

        # 충분히 스크롤을 내리기 위한 루프
        last_height = driver.execute_script("return document.body.scrollHeight")
        time.sleep(2)
        
        while True:
            # 스크롤을 내립니다
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # 페이지 로딩을 기다립니다
            time.sleep(2)
            # 새로운 스크롤 높이를 얻습니다
            new_height = driver.execute_script("return document.body.scrollHeight")
            # 더 이상 스크롤 높이가 변하지 않으면 루프를 종료합니다
            if new_height == last_height:
                break
            last_height = new_height

        # CSS 선택자를 사용하여 원하는 요소들을 찾기
        if day != 'complete':
            elements = driver.find_elements(By.CSS_SELECTOR, '#root > main > div > div.px-11.mx-auto.my-0.w-full.lg\\:w-default-max-width.md\\:w-\\[490px\\] > div.flex.flex-col.gap-4 > div.flex.flex-wrap.gap-4.content-start > div > div > div > a')
        else :
            elements = driver.find_elements(By.CSS_SELECTOR, '#root > main > div > div.px-11.mx-auto.my-0.w-full.lg\:w-default-max-width.md\:w-\[490px\] > div.flex.flex-wrap.gap-4.content-start > div > div > div > a')
        
            
        # 각 요소에서 href 추출하여 리스트에 저장
        for index, element in enumerate(elements):
            rank_list.append(index+1)
            day_list.append(day)
            href = element.get_attribute('href')
            if href:
                link_list.append(href)
                print(href)

        # 링크 리스트 확인
        print(f'총 {len(link_list)}개의 링크가 저장되었습니다.')
    
    
    
    day ='complete'

    
    driver.quit()
    
    data = {
        'Link': link_list,
        'Day': day_list,
        'Rank': rank_list
    }
    
    df_links = pd.DataFrame(data)
    return df_links
```

```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def get_webtoon_details(df_links):
    

    options = Options()
    user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36"
    options.add_argument('user-agent=' + user_agent)
    options.add_argument("lang=ko_KR")
    # options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")
    options.add_argument("--no-sandbox")
    driver = webdriver.Chrome(options=options)

    title_list = []
    genre_list = []
    view_list = []
    like_list = []
    story_list = []
    keywords_list = []
    age_list = []

    for index, row in df_links.iterrows():
        link = row['Link']
        try:
            driver.get(link)
            time.sleep(2) 
            
            # 로그인 페이지 확인 및 로그인 처리
            if "로그인이 필요합니다" in driver.page_source:
                kakao_id = KAKAO_ID # 환경 변수에서 카카오 ID 가져오기
                kakao_pw = KAKAO_PW # 환경 변수에서 카카오 PW 가져오기
                login_button = driver.find_element(By.CSS_SELECTOR, 'button.relative.px-10.py-0.btn-transparent-grey-06')
                login_button.click()  # '카카오계정 직접 입력' 버튼 클릭
                time.sleep(2)
                driver.find_element(By.ID, 'loginId--1').send_keys(kakao_id)
                driver.find_element(By.ID, 'password--2').send_keys(kakao_pw)
                driver.find_element(By.CSS_SELECTOR, 'button.btn_g.highlight.submit').click()

                time.sleep(5)  # 로그인 후 대기
                driver.implicitly_wait(10)
            

            wait = WebDriverWait(driver, 20)
            button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#root > main > div > div > div.inset-x-auto.relative > div.relative.-mb-11.z-2 > div.w-full.flex-center.sticky.top-header-height.z-4.h-sub-tab-height-v4 > div.w-full > div > div.w-full.h-full.relative > ul > li:nth-child(2)')))
            button.click()
            print("버튼이 클릭되었습니다.")

            info_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#root > main > div > div > div.relative.z-1.h-87 > div.pt-16.px-11')))
            info = info_element.text.split('\n')
            print(info)
            
            title_list.append(info[0] if len(info) > 0 else "")
            genre_list.append(info[2] if len(info) > 2 else "")
            view_list.append(info[3] if len(info) > 3 else "")
            like_list.append(info[4] if len(info) > 4 else "")

            plot_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#root > main > div > div > div.inset-x-auto.relative > div.relative.-mb-11.z-2 > div.h-full.mx-11.pb-100 > div > div:nth-child(2)')))
            story_list.append(plot_element.text)

            keywords_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#root > main > div > div > div.inset-x-auto.relative > div.relative.-mb-11.z-2 > div.h-full.mx-11.pb-100 > div > div:nth-child(3)')))
            keywords_list.append(keywords_element.text[1:])
            


            try:
                # Wait for the element to be present
                element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#root > main > div > div > div.inset-x-auto.relative > div.relative.-mb-11.z-2 > div.h-full.mx-11.pb-100 > div > div:nth-child(1) > div')))

                # Find child elements
                child_elements = element.find_elements(By.XPATH, './*')

                # Get the text of the element
                age_text = element.text

                # Preserved phrases list
                preserved_phrases = ["3시간마다 무료", "8시간마다 무료", "12시간마다 무료", 
                                     "1일마다 무료", "2일마다 무료", "3일마다 무료", "7일마다 무료"]

                preserved_phrase_found = None
                for phrase in preserved_phrases:
                    if phrase in age_text:
                        preserved_phrase_found = phrase
                        break

                if preserved_phrase_found:
                    # If a preserved phrase is found, split the text around it
                    parts = age_text.split(preserved_phrase_found)
                    before_preserved = parts[0].split()
                    after_preserved = parts[1].split() if len(parts) > 1 else []
                    age_split = before_preserved + [preserved_phrase_found] + after_preserved
                else:
                    # If no preserved phrase is found, split the whole text
                    age_split = age_text.split()

                block_len = len(age_split)

                if len(child_elements) != block_len:
                    age = '성인'
                elif '15' in age_split:
                    age = '15'
                else:
                    age = '전체'

            except Exception as e:
                print(f'age를 찾을 수 없습니다. 오류: {e}')
                   
            age_list.append(age)
            print(age)

            time.sleep(1)
            
        except Exception as e:
            print(f"Error occurred for link: {link}, Error: {e}")
            driver.save_screenshot(f"error_screenshot_{index}.png")
        driver.implicitly_wait(10)
        time.sleep(2) 
        
    driver.quit()
    
    # 모든 리스트가 동일한 길이를 가지도록 조정
    max_len = max(len(title_list), len(genre_list), len(view_list), len(like_list), len(story_list), len(keywords_list), len(age_list))
    
    def pad_list(lst, length):
        return lst + [''] * (length - len(lst))
    
    title_list = pad_list(title_list, max_len)
    genre_list = pad_list(genre_list, max_len)
    view_list = pad_list(view_list, max_len)
    like_list = pad_list(like_list, max_len)
    story_list = pad_list(story_list, max_len)
    keywords_list = pad_list(keywords_list, max_len)
    age_list = pad_list(age_list, max_len)
    
    data = {
        'Title': title_list,
        'Genre': genre_list,
        'View': view_list,
        'Story': story_list,
        'Keywords': keywords_list,
        'Likes': like_list,
        'Age': age_list
    }
    data = pd.DataFrame(data)
    df_details = pd.DataFrame(data)
    return df_details

```
