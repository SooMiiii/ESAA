# 04-2 matplotlib 라이브러리 자유자재로 사용하기

## 기초 그래프 그리기

```python
# 데이터 집합 불러오기
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
# 히스토그램 -> 데이터 분포와 빈도를 살펴보는 용도
# 변수를 하나만 사용해서 그린 그래프 : 일변량 그래프
fig = plt.figure() # 기본 틀 마련
axes1 = fig.add_subplot(1,1,1) # 격자 구성
axes1.hist(tips['total_bill'], bins=10) # 히스토그램 그리기
axes1.set_title('Histogram of Total Bill') # 그래프 제목 추가
axes1.set_xlabel('Frequency') # x축 제목 추가
axes1.set_ylabel('Total Bill') # y축 제목 추가
```

```python
# 산점도 그래프 그리기
# 변수 2개를 사용하기 때문에 이변량 그래프라고도 함
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Sxatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 박스 그래프 그리기
# 이산형 변수와 연속형 변수를 함께 사용하는 그래프.
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']],
              labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

# 다변량 그래프 그리기
- 3개 이상의 변수를 사용한 그래프

```python
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha = 0.5
)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# 04-3 seaborn 라이브러리 자유자재로 사용하기

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
# subplot,distplot 메서드 사용
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# 밀집도 그래프 제외 -> kde 인잣값을 False
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 밀집도 그래프만 나타내기 : hist 인자를 False
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# rug 인자를 추가하기
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
# count 그래프 : 이산값을 나타낸 그래프
ax = plt.subplots()
ax = sns.countplot(x='day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Dau of the Week')
ax.set_ylabel('Frequency')
```

### 다양한 종류의 이변량 그래프 그리기

#### 다양한 종류의 이변량 그래프 그리기
- seaborn 라이브러리의 rgplot -> 산점도 그리기, 회귀선도 그릴 수 있다.
- 회귀선 제거 : fig_reg 인자를 False


```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill',y='tip',data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

- jointplot 메서드 : 산점도 그래프와 히스토그램을 한 번에 그려준다.

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

- 육각 그래프
- 산점도 그래프는 겹쳐있는 점을 구분하기 어렵다는 단점이 있음. -> 육각 그래프 사용
- 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
- 데이터 개수가 많아지면 진한 색으로 표현
- jointplot 메서드를 그대로 사용, kind 대신 hex 로 인잣값 지정

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

- 이차원 밀집도 그리기 -> kdeplot 메서드로 만들 수 있다.

```python
ax=plt.subplots()
ax=sns.kdeplot(x=tips['total_bill'], y=tips['tip'], shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

- 관계 그래프
- 중복된 정보가 표현된다는 단점이 있다.
- map_upper 메서드는 대각선을 기준으로 위쪽에 그릴 그래프 지정
- mpa_diag 는 대각선을 중심으로 그래프를 그린다.

```python
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)  # 산점도 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)  # 이차원 밀집도
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) # 히스토그램, 밀집도
plt.show()

# 대각선 기준으로 위쪽에는 산점도, 아래쪽에는 이차원 밀집도, 대각선을 줌심으로 히스톡램과 밀집도
```

# 04-5 seaborn 라이브러리로 그래프 스타일 설정하기
- set_style 메서드 사용
- 실행하면 이후의 그래프가 모두 지정한 스타일로 바뀐다.

```python
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('whitegrid')
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
fig = plt.figure()

seaborn_styles=['darkgrid', 'whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
